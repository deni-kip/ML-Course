## Matplotlib Example Workflow

```python
# 0. Import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup Plots
fig, ax = plt.subplots(figsize=(10, 5))

# 3. Plot data
ax.plot(x, y)

# 4. customixe plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save and show the whole figure
fig.savefig('images/simple-plot.png')
```

## Two options for subplots


```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                              ncols=2,
                                              figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                  ncols=2,
                  figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/3);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matplotlib object oriented method)

* When plotting something quickly, it is okay to use pyplot method
* When plotting something more advance, use the oo method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            cmap='winter');
```

```python
# Object Oriented method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax,
            cmap='winter');
#ax.set_ylim([100, 700])
#ax.set_xlim([45, 100])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter');

# Customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
     xlabel='Age',
     ylabel='Cholestrol');

# Add a legend
ax.legend(*scatter.legend_elements(), title='target');

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
# Subplots of chol, age, and Thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target']);

# Customize the plot
ax0.set(title='Heart Disease and Cholesterol Levels',
#         xlabel='Age',
        ylabel='Cholestrol');

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title='target');

# Add a meanline
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target']);

# customize the plot
ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='thalach');

# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title='target');

# Adding a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

## Customizing the x and y axis limitation

```python
# Set the DataFrame
plt.style.use('ggplot')
```

```python
# Subplots of chol, age, and Thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter');

# Customize the plot
ax0.set(title='Heart Disease and Cholesterol Levels',
#         xlabel='Age',
        ylabel='Cholestrol');

# Setting the limits for ax0 x axis
ax0.set_xlim([50, 80])

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title='target');

# Add a meanline
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap='winter');

# customize the plot
ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='thalach');

# Setting the limits for ax1 x axis
ax1.set_xlim([50, 80])
# Setting the limits for ax1 y axis
ax1.set_ylim([60, 200])

# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title='target');

# Adding a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
